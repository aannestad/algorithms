```python
n = 10
```

```python
arr = [i for i in range(0,n)]
print(arr)
```

```python
from random import shuffle
arr2 = arr.copy()
shuffle(arr2)
print(arr2)
```

```python
# Deranged?
der = [i for i,j in zip(arr,arr2) if i==j]
if len(der) == 0:
    print("Derangement")
else:
    print(der)
```

```python
def derange(arr):
    arr2 = arr.copy()
    shuffle(arr2)
    der = [i for i,j in zip(arr,arr2) if i==j]
    return arr2, der
```

```python
d = 0
iters = 1000    # Select iterations
for i in range(1,iters):
    new_arr, stat_points = derange(arr)
    # print_arr(new_arr, stat_points)
    if len(stat_points) == 0:
        d += 1
print("Of " + str(iters) + " runs, " + str(d) + " were derangements")

print("Thereby, the probability of at least one ball in original position was: " + str(1-d/iters))
```

```python
import matplotlib.pyplot as plt

# Select grid size 3x3
n = 3

# Preparing dataset
x = [x % n for x in range(n**2)]
h = n
y = []
for i in range(n**2):
    if i % n == 0: # end of line
        h -= 1     # move up
    y.append(h)

balls = [x for x in range(1,n**2+1)]

# plotting scatter plot
#plt.figure(figsize=(5, 5))
plt.scatter(x, y,2000/n)
ax = plt.gca()
ax.set_aspect("equal")

# Loop for annotation of all points

for i in range(len(x)):
    plt.annotate(balls[i], (x[i]-0.01, y[i] - 0.01))

# adjusting the scale of the axes
plt.axis('off')
plt.xlim((-1,n))
plt.ylim((-1,n))
plt.show()
```

```python
ballsd = derange(balls)
marks = ballsd[1]
balls = ballsd[0]

plt.scatter(x, y,2000/n)

for i in range(len(x)):
    plt.annotate(balls[i], (x[i]-0.01, y[i] - 0.01))

ax = plt.gca()
ax.set_aspect("equal")

plt.axis('off')
plt.xlim((-1,n))
plt.ylim((-1,n))
for m in marks:
    y_adj = 0
    if m % n == 0 & m < n**2:
        y_adj = 1
    m1=plt.Circle((x[m-1], y[m-y_adj]), 0.3, color='r', fill=False)
    ax.add_patch(m1)
plt.show()
```
